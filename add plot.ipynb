```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties, fontManager
font_path = '/data/wangyuhang/ttf/Times New Roman.ttf'
font_prop = FontProperties(fname=font_path)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
```

```python
df=pd.read_csv('pans/0606-df_user_nation_treat_3m.csv',index_col=0)
df['ln_Score']=np.log(1+df['Score'])
df
```

```python
treat_mapping = {
    0: 'Control Group',
    1: 'Treatment Group'
}
after_mapping = {
    0: 'Before Treatment',
    1: 'After Treatment'
}
median_Q_Tags_len_filter_mapping={
   0: 'High Knowledge Complexity',
    1: 'Low Knowledge Complexity' 
}

df['treat'] = df['treat'].map(treat_mapping)
df['after'] = df['after'].map(after_mapping)
df['median_Q_Tags_len_filter'] = df['median_Q_Tags_len_filter'].map(median_Q_Tags_len_filter_mapping)
df
```

```python
avg_scores = df.groupby(['version_User_Id', 'treat', 'after', 'median_Q_Tags_len_filter']).agg(ln_Score=('ln_Score','mean')).reset_index()

avg_scores_1 = df[df['median_Q_Tags_len_filter']=='Low Knowledge Complexity'].groupby(['version_User_Id', 'treat', 'after']).agg(ln_Score=('ln_Score','mean')).reset_index()

avg_scores_2= df[df['median_Q_Tags_len_filter']=='High Knowledge Complexity'].groupby(['version_User_Id', 'treat', 'after']).agg(ln_Score=('ln_Score','mean')).reset_index()
avg_scores
```

```python
high_complexity = avg_scores[avg_scores['after'] == 'Before Treatment']
# 筛选低复杂度分组的数据
low_complexity = avg_scores[avg_scores['after'] == 'After Treatment']

# 
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='treat', y='ln_Score', hue='median_Q_Tags_len_filter', data=high_complexity)
plt.title('Before Treatment', fontproperties=font_prop,fontsize=16)
plt.xlabel('Group', fontproperties=font_prop,fontsize=16)
plt.ylabel('Average ln(Score)', fontproperties=font_prop,fontsize=16)
plt.legend(title='Knowledge Complexity', prop=font_prop,fontsize=16)
plt.ylim(0, 3)  # 



# 
plt.subplot(1, 2, 2)
sns.boxplot(x='treat', y='ln_Score', hue='median_Q_Tags_len_filter', data=low_complexity)
plt.title('After Treatment', fontproperties=font_prop,fontsize=16)
plt.xlabel('Group', fontproperties=font_prop,fontsize=16)
plt.ylabel('Average ln(Score)', fontproperties=font_prop,fontsize=16)
plt.legend(title='Knowledge Complexity', prop=font_prop,fontsize=16)
plt.ylim(0, 3)  #

# 
plt.tight_layout()
plt.show()
```
